```python
df = pd.read_csv('2_fields_new.csv')
df.head()
```

```python
#creating a list of all the response times so we can find the average
class_avg_resp_time = []

for num in range(1,6):
  class_avg_resp_time.append(df.loc[df['ques_no'] == num]['Response Time (in secs)'].mean())

class_avg_resp=class_avg_resp_time * 10

df.insert(loc=3, column='class average response time', value=class_avg_resp)
```

```python
df
```

```python
#creating a function that groups response time behaviour into 1 of three classes, pretty simple
def func(x, y):
  if x > y + 2:
    return 'MORE than average'
  elif x + 2 < y:
    return 'LESS than average'
  else:
    return 'Normal'
```

##### taking 2 seconds as the time error, if the difference b/w response time and class avg is +-2, we take the behaviour as normal.

```python
df['resp_behaviour(2 secs)'] = df.apply(lambda x: func(x['Response Time (in secs)'], x['class average response time']), axis=1)
```
